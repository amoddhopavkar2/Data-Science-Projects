# Scatter Plots

```python
import numpy as np
x = np.array([1, 2, 3, 4, 5, 6])
y = x

plt.figure()
plt.scatter(x, y)
```

```python
x = np.array([1, 2, 3, 4, 5, 6])
y = x

colors = []
for i in range(len(x)):
    if i % 2 == 0:
        colors.append('green')
    else:
        colors.append('red')

plt.figure()
plt.scatter(x, y, s = 50, c = colors)
```

# Line Plots

```python
import numpy as np
x = np.array([1, 2, 3, 4, 5, 6, 7, 8])
y = x ** 2

plt.figure()
plt.plot(x, '-o', y, '-o')
```

```python
plt.xlabel('X Index')
plt.ylabel('Y Index')
plt.title('X vs Y')

plt.legend(['X', 'Y'])
```

```python
plt.plot([10,20,30,40,50],'--r')
```

```python
plt.legend(['X', 'Y', 'Competition'])
```

```python
plt.gca().fill_between(range(len(x)),
                      x, y,
                      facecolor = 'blue',
                      alpha = 0.25)
```
