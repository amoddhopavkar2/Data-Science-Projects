```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib notebook
```

```python
plt.style.use('seaborn-colorblind')
```

```python
snow_df = pd.read_csv('snowfall.csv')
snow_df.head()
```

```python
temp_df = pd.read_excel('Temp_1961_1990.xls')
temp_df = temp_df.rename(columns = {' Month' : 'Month'})
temp_df.head()
```

```python
rain_df = pd.read_excel('Rain_1961_1990.xls')
rain_df = rain_df.rename(columns = {' Month' : 'Month'})
rain_df.head()
```

```python
Jan = snow_df['Jan'].mean()
Feb = snow_df['Feb'].mean()
Mar = snow_df['Mar'].mean()
Apr = snow_df['Apr'].mean()
May = snow_df['May'].mean()
Jun = snow_df['Jun'].mean()
Jul = snow_df['Jul'].mean()
Aug = snow_df['Aug'].mean()
Sep = snow_df['Sep'].mean()
Oct = snow_df['Oct'].mean()
Nov = snow_df['Nov'].mean()
Dec = snow_df['Dec'].mean()
```

```python
weather_df = pd.DataFrame({'Snowfall' : [Jan, Feb, Mar, Apr, May, Jun, Jul, Aug, Sep, Oct, Nov, Dec]})
weather_df
```

```python
monthly_temp = temp_df.groupby('Month').mean()['tas']
Jan = monthly_temp.iloc[0]
Feb = monthly_temp.iloc[1]
Mar = monthly_temp.iloc[2]
Apr = monthly_temp.iloc[3]
May = monthly_temp.iloc[4]
Jun = monthly_temp.iloc[5]
Jul = monthly_temp.iloc[6]
Aug = monthly_temp.iloc[7]
Sep = monthly_temp.iloc[8]
Oct = monthly_temp.iloc[9]
Nov = monthly_temp.iloc[10]
Dec = monthly_temp.iloc[11]
```

```python
weather_df['Temperature'] = [Jan, Feb, Mar, Apr, May, Jun, Jul, Aug, Sep, Oct, Nov, Dec]
weather_df
```

```python
monthly_rain = rain_df.groupby('Month').mean()['pr']
monthly_rain.iloc[0]
Jan = monthly_rain.iloc[0]
Feb = monthly_rain.iloc[1]
Mar = monthly_rain.iloc[2]
Apr = monthly_rain.iloc[3]
May = monthly_rain.iloc[4]
Jun = monthly_rain.iloc[5]
Jul = monthly_rain.iloc[6]
Aug = monthly_rain.iloc[7]
Sep = monthly_rain.iloc[8]
Oct = monthly_rain.iloc[9]
Nov = monthly_rain.iloc[10]
Dec = monthly_rain.iloc[11]
```

```python
weather_df['Rainfall'] = [Jan, Feb, Mar, Apr, May, Jun, Jul, Aug, Sep, Oct, Nov, Dec]
weather_df
```

```python
weather_df['Snowfall'] = weather_df['Snowfall'] / 1000
weather_df
```

```python
weather_df.index = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
```

```python
weather_df
```

```python
X = weather_df.index.tolist()
Y1 = weather_df['Snowfall'].tolist()
Y2 = weather_df['Temperature'].tolist()
Y3 = weather_df['Rainfall'].tolist()

plt.plot(X, Y1, '-s', label = "Snowfall (in 0.1*m)")
plt.plot(X, Y2, '-*', label = "Temperature (in $\,^{\circ}\mathrm{C}$)")
plt.plot(X, Y3, '-o', label = "Rainfall (in 0.1*cm)")

plt.title('Average Temperature, Snowfall and Rainfall in United States(1961-1990)')
plt.legend()
```
