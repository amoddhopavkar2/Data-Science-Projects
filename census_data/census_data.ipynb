# Census Data

Perform the following opertaions on US Beaureu Census data, which contains the population estimates for various counties.

```python
import pandas as pd
import numpy as np
census_df = pd.read_csv('census.csv')
census_df.head(10)
```

### Which state has the most counties in it?

```python
def max_counties():
    df = census_df[census_df['SUMLEV'] == 50]
    df = df.groupby(['SUMLEV', 'STNAME']).size().to_frame(name = 'count').reset_index()
    return df.loc[df['count'].idxmax()]['STNAME']
max_counties()
```

### The 3 most populas counties for each state

Arrange the 3 most populas counties in descending order(highest to lowest). Use CENSUS2010POP

```python
def most_populas_counties():
    df = census_df[census_df['SUMLEV'] == 50]
    df1 = df.sort_values(['STNAME', 'CENSUS2010POP'], ascending = False).groupby('STNAME').head(3).copy()
    df2 = df1.reset_index().groupby('STNAME').sum().sort_values(['CENSUS2010POP'], ascending = False).head(3).copy()
    return list(df2.index.values)
most_populas_counties()
```

### Which county has the largest change of population between 2010 to 2015?

Check for absoulute largest change. 

```python
def max_std_deviation():
    df = census_df[census_df['SUMLEV'] == 50]
    df['STDDEV'] = df[['POPESTIMATE2010', 'POPESTIMATE2011', 'POPESTIMATE2012', 
                     'POPESTIMATE2013', 'POPESTIMATE2014', 'POPESTIMATE2015']].std(axis = 1)
    return df.loc[df['STDDEV'].idxmax()]['CTYNAME']
max_std_deviation()
```

### The county belongs to Region 1 or Region 2 and starts with 'Washington'

The United States is divided into 4 regions. Find a county which belongs to either Region 1 or Region 2, with it's name starting with 'Washington' and the POPESTIMATE2015 > POPESTIMATE2014 

```python
def query():
    df = census_df[census_df['SUMLEV'] == 50]
    region_1 = df['REGION'] == 1
    region_2 = df['REGION'] == 2
    county_name = df['CTYNAME'].str.startswith('Washington')
    population_estimate = df['POPESTIMATE2015'] > df['POPESTIMATE2014']
    df = df[(region_1 | region_2) & county_name & population_estimate]
    return df[['STNAME', 'CTYNAME']]
query()
```

### Average Population in the counties

Calculate the average population according to CENSUS2010POP for the counties of each state using the groupby() function. Trim the decimals upto 2 places

```python
def avg_population():
    for group, frame in census_df.groupby('STNAME'):
        avg = np.average(frame['CENSUS2010POP'])
        print('Counties in the state of {} have an average population of: {:.2f}'.format(group,avg))
avg_population()
```
