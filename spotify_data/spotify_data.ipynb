# Spotify

Sort the songs on the basis of popularity, and represent the top 5 and bottom 5 songs. 

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('top50.csv')
df = df.rename(columns = {'Track.Name' : 'Track'})
df.head()
```

```python
df = df[:8]
```

```python
plt.scatter(df['Track'], df['Popularity'], color = 'red')
plt.xticks(rotation = 45)
plt.subplots_adjust(bottom = 0.5)
```
